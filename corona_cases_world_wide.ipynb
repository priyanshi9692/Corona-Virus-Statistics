```python
#selecting the columns required
df = df[['Places reporting cases','Cases']]
#renaming the column name
data_for_map = df.rename(index=str, columns={'Places reporting cases': 'CNTRY_NAME','Cases': 'Cases'})
```

```python
data_for_map.head()
```

```python
# joining the geodataframe with the cleaned up csv dataframe
merged = map_df.set_index('CNTRY_NAME').join(data_for_map.set_index('CNTRY_NAME'))
#.head() returns the top 5(by default ) lines of the dataframe
merged.head()
```

```python
merged=merged.fillna(0)
merged
```

```python
# set a variable that will call whatever column we want to visualise on the map
variable = 'Cases'
# set the range for the choropleth
vmin, vmax = 0, 80000
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(20, 8))
# remove the axis
ax.axis('off')
# add a title
ax.set_title('Corona Cases World Wide', fontdict={'fontsize':'25', 'fontweight' : '3'})
# create an annotation for the data source
ax.annotate('Source: WorldOMeter', xy=(0.1, .08), 
            xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', 
            fontsize=12, color='#555579')
sm = plt.cm.ScalarMappable(cmap='Oranges', norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
merged.plot(column=variable, cmap='Oranges', linewidth=1, ax=ax, edgecolor='0.5')
```

```python
fig.savefig('map_export.png', dpi=300)
```
